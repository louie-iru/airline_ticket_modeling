```python
import requests
import json
import pandas as pd
from tqdm import tqdm, trange

pd.set_option('display.max_columns', None)
```

```python
service_key = "OL3hJa7Y++gNZQ3Y13Ap8morR8uR1hBd6ajFJLup4xmfMekt3Tn7h6JWVclkx/bkVAJG92ZNGqgeAfsg3Bjlcw=="
url = "https://apis.data.go.kr/1360000/AirInfoService/getAirInfo"

start_date = "20220101"
end_date = "20230401"

air_df = pd.DataFrame()

for date in tqdm(pd.date_range(start=start_date, end=end_date)):
    parameters = {
        "serviceKey": service_key,
        "numOfRows": "1",
        "pageNo": "1",
        "dataType": "JSON",
        "fctm": date.strftime("%Y%m%d"),
        "icaoCode": "RKPC"
    }
    response = requests.get(url, params=parameters)
    data = json.loads(response.content)

    if data["response"]["header"]["resultCode"] == "00":
        item = data["response"]["body"]["items"]["item"]
        air_df = pd.concat([air_df, pd.DataFrame(item)])

air_df = air_df.reset_index(drop=True)
air_df = air_df.astype({
'icaoCode': 'string'
, 'airportName': 'string'
, 'tmFc': 'datetime64'
, 'wd': 'int64'
, 'ws': 'int64'
, 'ta': 'int64'
, 'qnh': 'int64'
})
```
