```python
import pandas as pd
```

```python
dataset = pd.read_csv("./data/AIRLINE_TICKET_2022.csv", encoding = 'cp949')

display(dataset, dataset.info())
```

```python
dataset = dataset.astype({
    'DEPART_DATE': 'datetime64'
    , 'DEPART_TIME': 'string'
    , 'AIRLINE': 'category'
    , 'FLIGHT_NO': 'string'
    , 'DEPART': 'category'
    , 'ARRIVE': 'category'
    , 'CLASS_CODE': 'category'
    , 'CLASS_DESC': 'category'
    , 'AVAIL_SEAT': 'int64'
    # , 'FARE_ORIGIN': 'int64' # Null값 존재 + KE에서만 유효하다.
    , 'FARE': 'int64'
    , 'SEARCH_DATE': 'datetime64'
    , 'SEARCH_TIME': 'string'
})

display(dataset, dataset.info())
```

```python
dataset.to_csv('./data/AIRLINE_TICKET_2022_2.csv', index = False, encoding = 'utf-8-sig')
```
