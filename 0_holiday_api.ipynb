```python
import requests
import xml.etree.ElementTree as ET
import pandas as pd
```

```python
service_key = "OL3hJa7Y++gNZQ3Y13Ap8morR8uR1hBd6ajFJLup4xmfMekt3Tn7h6JWVclkx/bkVAJG92ZNGqgeAfsg3Bjlcw=="

url = "http://apis.data.go.kr/B090041/openapi/service/SpcdeInfoService/getHoliDeInfo"

dataframes = []

for year in range(2022, 2024):
    for month in range(1, 13):
        parameters = {
            "solYear" : str(year)
            , "solMonth" : str(month).zfill(2)
            , "ServiceKey" : service_key
        }

        response = requests.get(url, params = parameters)

        root = ET.fromstring(response.content)
        items = root.find('body').find('items').findall('item')

        data = []
        for item in items:
            date = item.find('locdate').text
            name = item.find('dateName').text
            is_holiday = item.find('isHoliday').text
            data.append([date, name, is_holiday])

        df = pd.DataFrame(data, columns=['DT', 'DESC', 'HOLIDAY'])
        dataframes.append(df)

    result = pd.concat(dataframes, ignore_index=True)
    
result = result.astype({
    "DT": "datetime64"
    , "DESC": "string"
    , "HOLIDAY": "category"
})
result
```

```python
result.to_csv('./data/HOLIDAY_2022.csv', index=False)
```
